```python
# =============================================================================
#  PROTON AFFINITY ML — COMPLETE TRAINING PIPELINE
#  Save as PA_TRAINER.ipynb in Google Colab
#  Run once → saves model artifacts → use PA_PREDICTOR.ipynb for new molecules
# =============================================================================

# ─────────────────────────────────────────────────────────────────────────────
# CELL 1 ─ Install dependencies
# ─────────────────────────────────────────────────────────────────────────────
import subprocess
subprocess.run(["pip", "install", "-q", "rdkit", "xgboost", "shap",
                "scikit-learn", "pandas", "matplotlib", "seaborn", "openpyxl"],
               check=True)
print("✓ Dependencies installed")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 2 ─ Imports
# ─────────────────────────────────────────────────────────────────────────────
import json, warnings, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import shap
import xgboost as xgb
import joblib

from rdkit import Chem
from rdkit.Chem import Descriptors, MACCSkeys, AllChem, Descriptors3D

from sklearn.ensemble import (RandomForestRegressor, GradientBoostingRegressor,
                               ExtraTreesRegressor, VotingRegressor, StackingRegressor)
from sklearn.linear_model  import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline      import Pipeline
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics        import mean_absolute_error, r2_score, mean_squared_error

warnings.filterwarnings('ignore')
print("✓ All imports successful")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 3 ─ Load data  (upload your .csv or .xlsx when prompted)
# ─────────────────────────────────────────────────────────────────────────────
from google.colab import files

print("Please upload your dataset file (.csv or .xlsx)...")
uploaded = files.upload()
filename = list(uploaded.keys())[0]
print(f"Uploaded: {filename}")

if filename.endswith('.csv'):
    df = pd.read_csv(filename)
elif filename.endswith(('.xls', '.xlsx')):
    df = pd.read_excel(filename)
else:
    raise ValueError("Unsupported format — please upload a .csv or .xlsx file")

# Expected columns: 'SMILES' and 'PA(kcal/mol)'
print(f"Loaded: {df.shape[0]} rows × {df.shape[1]} columns")
print(df.head())

# ─────────────────────────────────────────────────────────────────────────────
# CELL 4 ─ Clean & label
# ─────────────────────────────────────────────────────────────────────────────
df['mol'] = df['SMILES'].apply(Chem.MolFromSmiles)
df        = df[df['mol'].notna()].copy()

# Deduplicate on canonical SMILES
df['can_smi'] = df['mol'].apply(Chem.MolToSmiles)
df = df.drop_duplicates(subset='can_smi').reset_index(drop=True)

df['is_sponge'] = (df['PA(kcal/mol)'] > 240).astype(int)

print(f"Clean molecules : {len(df)}")
print(df['is_sponge'].value_counts().rename({0: 'Normal', 1: 'Proton sponge'}))

# ─────────────────────────────────────────────────────────────────────────────
# CELL 5 ─ Stratified train / test split  (80/20, locked before any fitting)
# ─────────────────────────────────────────────────────────────────────────────
train_idx, test_idx = train_test_split(
    df.index, test_size=0.20, random_state=42, stratify=df['is_sponge']
)
df_train = df.loc[train_idx].reset_index(drop=True)
df_test  = df.loc[test_idx].reset_index(drop=True)

print(f"Train: {len(df_train)}  |  Test (locked): {len(df_test)}")
print(f"Train sponges: {df_train['is_sponge'].sum()}  |  "
      f"Test sponges: {df_test['is_sponge'].sum()}")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 6 ─ Descriptor engine  (2D + 3D conformer + fingerprints + SMARTS flags)
# ─────────────────────────────────────────────────────────────────────────────
SMARTS_DEFS = {
    'amidine'     : '[NX3][CX3]=[NX2]',
    'guanidine'   : '[NX3][CX3](=[NX2])[NX3]',
    'peri_diamine': 'c1ccc2c(N)cccc2c1N',
    'bicyclic_N'  : '[N;R2]',
    'phosphazene' : '[NX2]=[PX4]([NX3])([NX3])[NX3]',
    'two_N_close' : '[#7]~[#7]',
    'vinamidine'  : '[NX3]/C=C/C=[NX2]',
}
SMARTS_COMPILED = {k: Chem.MolFromSmarts(v) for k, v in SMARTS_DEFS.items()}


def get_3d_descriptors(mol):
    """Generate 3D conformer (ETKDG + MMFF94) and compute shape descriptors."""
    keys = ['PMI1','PMI2','PMI3','Asphericity','Eccentricity',
            'NPR1','NPR2','RadiusOfGyration','InertialShapeFactor','SpherocityIndex']
    nan_row = {k: np.nan for k in keys}
    try:
        mol_h = Chem.AddHs(mol)
        params = AllChem.ETKDGv3()
        params.randomSeed = 42
        if AllChem.EmbedMolecule(mol_h, params) == -1:
            return nan_row
        res = AllChem.MMFFOptimizeMolecule(mol_h, maxIters=500)
        if res == -1:
            AllChem.UFFOptimizeMolecule(mol_h)
        return {
            'PMI1'               : Descriptors3D.PMI1(mol_h),
            'PMI2'               : Descriptors3D.PMI2(mol_h),
            'PMI3'               : Descriptors3D.PMI3(mol_h),
            'Asphericity'        : Descriptors3D.Asphericity(mol_h),
            'Eccentricity'       : Descriptors3D.Eccentricity(mol_h),
            'NPR1'               : Descriptors3D.NPR1(mol_h),
            'NPR2'               : Descriptors3D.NPR2(mol_h),
            'RadiusOfGyration'   : Descriptors3D.RadiusOfGyration(mol_h),
            'InertialShapeFactor': Descriptors3D.InertialShapeFactor(mol_h),
            'SpherocityIndex'    : Descriptors3D.SpherocityIndex(mol_h),
        }
    except Exception:
        return nan_row


def build_full_feature_row(mol):
    """Build the complete feature vector for one RDKit mol object."""
    feat = {}

    # 2D RDKit physicochemical descriptors
    for name, fn in Descriptors.descList:
        try:    feat[name] = fn(mol)
        except: feat[name] = np.nan

    # Morgan ECFP4 fingerprint (1024 bits)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)
    for i in range(1024):
        feat[f'FP_{i}'] = int(fp[i])

    # MACCS keys (167 bits)
    maccs = MACCSkeys.GenMACCSKeys(mol)
    for i in range(167):
        feat[f'MACCS_{i}'] = int(maccs[i])

    # SMARTS substructure flags
    for name, pat in SMARTS_COMPILED.items():
        feat[f'has_{name}'] = int(mol.HasSubstructMatch(pat)) if pat else 0

    # 3D conformer descriptors
    feat.update(get_3d_descriptors(mol))

    return feat


def build_feature_matrix(mol_series, desc="Building features"):
    """Vectorise build_full_feature_row over a pandas Series of mol objects."""
    rows = []
    n = len(mol_series)
    for i, mol in enumerate(mol_series):
        if (i + 1) % 50 == 0 or i == n - 1:
            print(f"  {desc}: {i+1}/{n}", end='\r')
        rows.append(build_full_feature_row(mol))
    print()
    return pd.DataFrame(rows)


print("Computing features for TRAIN set (this takes a few minutes)...")
X_train_raw = build_feature_matrix(df_train['mol'], "Train")

print("Computing features for TEST set...")
X_test_raw  = build_feature_matrix(df_test['mol'],  "Test")

print(f"Raw feature shape — Train: {X_train_raw.shape}  |  Test: {X_test_raw.shape}")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 7 ─ Feature selection & preprocessing
# ─────────────────────────────────────────────────────────────────────────────
# 1. Drop zero-variance columns (computed on train)
nonzero_cols = X_train_raw.columns[X_train_raw.std() > 0].tolist()
X_train_raw  = X_train_raw[nonzero_cols]
X_test_raw   = X_test_raw[nonzero_cols]

# 2. Impute with train median
train_median = X_train_raw.median()
X_train_raw  = X_train_raw.fillna(train_median)
X_test_raw   = X_test_raw.fillna(train_median)

# 3. Drop highly correlated features (train correlation matrix only)
corr_matrix  = X_train_raw.corr().abs()
upper        = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
drop_cols    = [c for c in upper.columns if (upper[c] > 0.95).any()]
X_train      = X_train_raw.drop(columns=drop_cols)
X_test       = X_test_raw.drop(columns=drop_cols)

print(f"Final feature matrix: {X_train.shape[1]} features")
print(f"  Dropped (zero-variance) : {len(nonzero_cols) - len(X_train_raw.columns) + len(drop_cols)}")
print(f"  Dropped (high corr>0.95): {len(drop_cols)}")

y_train = df_train['PA(kcal/mol)'].values
y_test  = df_test['PA(kcal/mol)'].values
g_train = df_train['is_sponge'].values
g_test  = df_test['is_sponge'].values

# Save preprocessing metadata
preprocessing_meta = {
    'nonzero_cols' : nonzero_cols,
    'drop_cols'    : drop_cols,
    'train_median' : train_median.to_dict(),
    'final_columns': X_train.columns.tolist(),
}
with open('preprocessing_meta.json', 'w') as f:
    json.dump(preprocessing_meta, f)
print("Saved: preprocessing_meta.json")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 8 ─ Model definitions
# ─────────────────────────────────────────────────────────────────────────────
voting_est = VotingRegressor(estimators=[
    ('xgb', xgb.XGBRegressor(
        n_estimators=500, learning_rate=0.05, max_depth=6,
        subsample=0.8, colsample_bytree=0.8, reg_lambda=1.0,
        min_child_weight=3, random_state=42, n_jobs=-1, verbosity=0)),
    ('et',  ExtraTreesRegressor(
        n_estimators=300, min_samples_leaf=2, random_state=42, n_jobs=-1)),
    ('lasso', Pipeline([('sc', StandardScaler()),
                        ('l',  Lasso(alpha=0.1, max_iter=10000))]))
], weights=[2, 1, 1])

stacking_est = StackingRegressor(
    estimators=[
        ('xgb', xgb.XGBRegressor(
            n_estimators=400, learning_rate=0.05, max_depth=6,
            subsample=0.8, colsample_bytree=0.8, random_state=42,
            n_jobs=-1, verbosity=0)),
        ('et',  ExtraTreesRegressor(
            n_estimators=300, min_samples_leaf=2, random_state=42, n_jobs=-1)),
        ('rf',  RandomForestRegressor(
            n_estimators=300, min_samples_leaf=2, random_state=42, n_jobs=-1)),
        ('lasso', Pipeline([('sc', StandardScaler()),
                            ('l',  Lasso(alpha=0.1, max_iter=10000))]))
    ],
    final_estimator=Ridge(alpha=1.0),
    cv=5, n_jobs=-1
)

models = {
    'XGBoost'           : xgb.XGBRegressor(
                              n_estimators=600, learning_rate=0.04, max_depth=6,
                              subsample=0.8, colsample_bytree=0.8,
                              min_child_weight=3, reg_alpha=0.1, reg_lambda=1.0,
                              random_state=42, n_jobs=-1, verbosity=0),
    'Gradient Boosting' : GradientBoostingRegressor(
                              n_estimators=300, learning_rate=0.05,
                              max_depth=5, random_state=42),
    'Random Forest'     : RandomForestRegressor(
                              n_estimators=300, min_samples_leaf=2,
                              random_state=42, n_jobs=-1),
    'ExtraTrees'        : ExtraTreesRegressor(
                              n_estimators=300, min_samples_leaf=2,
                              random_state=42, n_jobs=-1),
    'VotingRegressor'   : voting_est,
    'StackingRegressor' : stacking_est,
    'Lasso'             : Pipeline([('scaler', StandardScaler()),
                                    ('lasso',  Lasso(alpha=0.1, max_iter=10000))]),
}

# ─────────────────────────────────────────────────────────────────────────────
# CELL 9 ─ Stratified 5-fold cross-validation on TRAIN set only
# ─────────────────────────────────────────────────────────────────────────────
skf    = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_res = {}

print("=" * 85)
print(f"  {'Model':<22} {'R²':>7} {'±std':>6} {'RMSE':>7} {'MAE':>7} "
      f"{'MAE_normal':>11} {'MAE_sponge':>11}")
print("=" * 85)

for name, model in models.items():
    fold_r2, fold_rmse, fold_mae, fold_mae_n, fold_mae_s = [], [], [], [], []

    for tr, te in skf.split(X_train, g_train):
        Xf_tr, Xf_te = X_train.iloc[tr], X_train.iloc[te]
        yf_tr, yf_te = y_train[tr],      y_train[te]
        gf_te        = g_train[te]

        model.fit(Xf_tr, yf_tr)
        pred = model.predict(Xf_te)

        n_mask = gf_te == 0
        s_mask = gf_te == 1

        fold_r2.append(r2_score(yf_te, pred))
        fold_rmse.append(np.sqrt(mean_squared_error(yf_te, pred)))
        fold_mae.append(mean_absolute_error(yf_te, pred))
        fold_mae_n.append(mean_absolute_error(yf_te[n_mask], pred[n_mask]) if n_mask.any() else np.nan)
        fold_mae_s.append(mean_absolute_error(yf_te[s_mask], pred[s_mask]) if s_mask.any() else np.nan)

    cv_res[name] = {
        'R2'        : np.mean(fold_r2),
        'R2_std'    : np.std(fold_r2),
        'RMSE'      : np.mean(fold_rmse),
        'MAE'       : np.mean(fold_mae),
        'MAE_normal': np.nanmean(fold_mae_n),
        'MAE_sponge': np.nanmean(fold_mae_s),
    }
    r = cv_res[name]
    print(f"  {name:<22} {r['R2']:>7.4f} {r['R2_std']:>6.4f} "
          f"{r['RMSE']:>7.2f} {r['MAE']:>7.2f} "
          f"{r['MAE_normal']:>11.2f} {r['MAE_sponge']:>11.2f}")

print("=" * 85)

cv_df = pd.DataFrame(cv_res).T.round(4)
cv_df.to_csv('cv_results.csv')
print("Saved: cv_results.csv")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 10 ─ Consensus ranking & identify best model
# ─────────────────────────────────────────────────────────────────────────────
rank_df = pd.DataFrame(index=cv_df.index)
rank_df['Rank_R2']   = cv_df['R2'].rank(ascending=False).astype(int)
rank_df['Rank_RMSE'] = cv_df['RMSE'].rank(ascending=True).astype(int)
rank_df['Rank_MAE']  = cv_df['MAE'].rank(ascending=True).astype(int)
rank_df['Mean_Rank'] = rank_df[['Rank_R2','Rank_RMSE','Rank_MAE']].mean(axis=1)
rank_df = rank_df.sort_values('Mean_Rank')
rank_df['Final_Rank'] = range(1, len(rank_df) + 1)

print("\n  CONSENSUS RANKING (lower = better):\n")
print(rank_df.to_string())

best_model_name = rank_df.index[0]
print(f"\n  >>> Best model: {best_model_name} <<<")

# ── Plot CV performance ───────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('5-Fold Cross-Validation Performance (Train Set)', fontsize=13, fontweight='bold')

metrics = [('R2', 'R² (higher=better)', False),
           ('RMSE', 'RMSE kcal/mol (lower=better)', True),
           ('MAE',  'MAE kcal/mol (lower=better)',  True)]

for ax, (metric, label, lower_better) in zip(axes, metrics):
    vals   = cv_df[metric].sort_values(ascending=lower_better)
    colors = ['#27AE60' if i == 0 else '#3498DB' if i < 3 else '#95A5A6'
              for i in range(len(vals))]
    ax.barh(vals.index, vals.values, color=colors, alpha=0.85)
    ax.set_xlabel(label, fontsize=10)
    ax.set_title(metric, fontweight='bold')
    for i, (idx, val) in enumerate(vals.items()):
        ax.text(val * 1.005 if not lower_better else val + 0.02,
                i, f'{val:.4f}', va='center', fontsize=8)

plt.tight_layout()
plt.savefig('cv_performance.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: cv_performance.png")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 11 ─ Retrain best model on FULL TRAIN set → evaluate on locked TEST set
# ─────────────────────────────────────────────────────────────────────────────
print(f"\nRetraining {best_model_name} on full training set ({len(X_train)} molecules)...")
final_model = models[best_model_name]
final_model.fit(X_train, y_train)
print("  Done.")

y_pred_train = final_model.predict(X_train)
print(f"\n  Train R²  : {r2_score(y_train, y_pred_train):.4f}")
print(f"  Train MAE : {mean_absolute_error(y_train, y_pred_train):.2f} kcal/mol")
print(f"  Train RMSE: {np.sqrt(mean_squared_error(y_train, y_pred_train)):.2f} kcal/mol")

y_pred_test = final_model.predict(X_test)
test_r2   = r2_score(y_test, y_pred_test)
test_mae  = mean_absolute_error(y_test, y_pred_test)
test_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))

n_mask_test = g_test == 0CELL 12 ─ Predicted vs Actual scatter — top 3 models (held-out test set)
s_mask_test = g_test == 1

print(f"\n  ══ HELD-OUT TEST SET PERFORMANCE ({len(X_test)} molecules) ══")
print(f"  Overall  — R²: {test_r2:.4f}  MAE: {test_mae:.2f}  RMSE: {test_rmse:.2f} kcal/mol")
if n_mask_test.any():
    print(f"  Normal   — MAE: {mean_absolute_error(y_test[n_mask_test], y_pred_test[n_mask_test]):.2f} kcal/mol")
if s_mask_test.any():
    print(f"  Sponges  — MAE: {mean_absolute_error(y_test[s_mask_test], y_pred_test[s_mask_test]):.2f} kcal/mol")

# ─────────────────────────────────────────────────────────────────────────────
#
# ─────────────────────────────────────────────────────────────────────────────
top3 = rank_df.index[:3].tolist()

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Predicted vs Actual PA — Top 3 Models (Held-Out Test Set)\n'
             'Blue = Normal base  |  Red = Proton sponge',
             fontsize=13, fontweight='bold')

for ax, name in zip(axes, top3):
    m = models[name]
    m.fit(X_train, y_train)
    pred = m.predict(X_test)

    ax.scatter(y_test[n_mask_test], pred[n_mask_test],
               alpha=0.5, s=20, color='steelblue', label='Normal', zorder=2)
    ax.scatter(y_test[s_mask_test], pred[s_mask_test],
               alpha=0.9, s=55, color='tomato',    label='Proton sponge', zorder=3)

    lo, hi = y_test.min() - 3, y_test.max() + 3
    ax.plot([lo, hi], [lo, hi], 'k--', linewidth=1.2)

    r2_   = r2_score(y_test, pred)
    mae_  = mean_absolute_error(y_test, pred)
    rmse_ = np.sqrt(mean_squared_error(y_test, pred))
    rk    = rank_df.loc[name, 'Final_Rank']

    ax.set_title(f'#{rk} {name}\nR²={r2_:.4f}  MAE={mae_:.2f}  RMSE={rmse_:.2f}',
                 fontsize=10, fontweight='bold')
    ax.set_xlabel('Actual PA (kcal/mol)', fontsize=10)
    ax.set_ylabel('Predicted PA (kcal/mol)', fontsize=10)
    ax.legend(fontsize=8)

plt.tight_layout()
plt.savefig('test_set_scatter_top3.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: test_set_scatter_top3.png")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 13 ─ Feature importance — top 15 features of the BEST MODEL
#   Works for tree-based models (XGBoost, RF, ET, GB) via built-in importance.
#   For ensemble/stacking the XGBoost sub-model is used as proxy.
# ─────────────────────────────────────────────────────────────────────────────
print(f"\nExtracting feature importance from: {best_model_name}")

def get_feature_importances(model, model_name, feature_names):
    """
    Return a Series of feature importances (mean decrease in impurity /
    XGBoost gain) indexed by feature name.
    Handles single tree models and ensemble / stacking wrappers.
    """
    tree_attrs = ['feature_importances_']   # sklearn tree models

    # Direct tree model
    if hasattr(model, 'feature_importances_'):
        return pd.Series(model.feature_importances_, index=feature_names)

    # XGBRegressor — use 'gain' importance for best interpretability
    if hasattr(model, 'get_booster'):
        scores = model.get_booster().get_score(importance_type='gain')
        imp = pd.Series({f: scores.get(f, 0.0) for f in feature_names})
        return imp / imp.sum()    # normalise to [0,1]

    # Pipeline (e.g. Lasso) — check the final step
    if hasattr(model, 'named_steps'):
        final_step = list(model.named_steps.values())[-1]
        if hasattr(final_step, 'coef_'):
            return pd.Series(np.abs(final_step.coef_), index=feature_names)

    # VotingRegressor / StackingRegressor — try XGBoost sub-model first
    for attr in ['estimators_', 'named_estimators_']:
        if hasattr(model, attr):
            sub_models = getattr(model, attr)
            # named_estimators_ is a dict, estimators_ is a list of (name, est)
            if isinstance(sub_models, dict):
                sub_models = list(sub_models.items())
            for item in sub_models:
                name_sub, est = item if isinstance(item, tuple) else (None, item)
                if hasattr(est, 'feature_importances_'):
                    return pd.Series(est.feature_importances_, index=feature_names)
                if hasattr(est, 'get_booster'):
                    scores = est.get_booster().get_score(importance_type='gain')
                    imp = pd.Series({f: scores.get(f, 0.0) for f in feature_names})
                    return imp / imp.sum()

    raise ValueError(f"Cannot extract feature importances from: {model_name}")


feat_imp = get_feature_importances(final_model, best_model_name, X_train.columns.tolist())
top15_imp = feat_imp.nlargest(15).sort_values(ascending=True)  # sort for horizontal bar

# ── Plot 1: Horizontal bar chart — Top 15 feature importances ────────────────
fig, ax = plt.subplots(figsize=(10, 7))
colors_imp = plt.cm.RdYlGn(np.linspace(0.3, 0.9, 15))
bars = ax.barh(top15_imp.index, top15_imp.values,
               color=colors_imp, alpha=0.88, edgecolor='white', linewidth=0.6)

for bar, val in zip(bars, top15_imp.values):
    ax.text(bar.get_width() + top15_imp.values.max() * 0.01,
            bar.get_y() + bar.get_height() / 2,
            f'{val:.4f}', va='center', fontsize=9)

ax.set_xlabel('Feature Importance (normalised)', fontsize=12)
ax.set_title(f'Top 15 Feature Importances — {best_model_name}', fontsize=13, fontweight='bold')
ax.set_xlim(0, top15_imp.values.max() * 1.18)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('feature_importance_top15.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: feature_importance_top15.png")

# Save importance table
feat_imp.nlargest(15).reset_index().rename(
    columns={'index': 'Feature', 0: 'Importance'}
).to_csv('feature_importance_top15.csv', index=False)
print("Saved: feature_importance_top15.csv")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 14 ─ SHAP interpretability — best model, top 15 features, TEST SET
# ─────────────────────────────────────────────────────────────────────────────
print(f"\nComputing SHAP values for {best_model_name} on test set (top 15 features)...")

TREE_MODELS = ['XGBoost', 'Gradient Boosting', 'Random Forest', 'ExtraTrees']

def get_shap_explainer_and_values(model, model_name, X_bg, X_explain):
    """
    Return (explainer, shap_values) for the given model.
    Uses TreeExplainer for tree-based and XGBoost sub-models.
    Falls back to KernelExplainer for linear / ensemble wrappers.
    """
    # Direct tree model (sklearn or xgboost)
    if model_name in TREE_MODELS or hasattr(model, 'feature_importances_') \
            or hasattr(model, 'get_booster'):
        exp = shap.TreeExplainer(model)
        return exp, exp.shap_values(X_explain)

    # Try to find a tree-based sub-model inside ensemble / stacking
    for attr in ['named_estimators_', 'estimators_']:
        if hasattr(model, attr):
            sub_models = getattr(model, attr)
            if isinstance(sub_models, dict):
                sub_models = list(sub_models.items())
            for item in sub_models:
                _, est = item if isinstance(item, tuple) else (None, item)
                if hasattr(est, 'get_booster') or hasattr(est, 'feature_importances_'):
                    print(f"  (Using tree sub-model for SHAP)")
                    exp = shap.TreeExplainer(est)
                    return exp, exp.shap_values(X_explain)

    # Fallback: KernelExplainer (slow, capped at 200 samples)
    print("  (Using KernelExplainer — this may be slow)")
    bg  = shap.sample(X_bg, 100, random_state=42)
    exp = shap.KernelExplainer(model.predict, bg)
    return exp, exp.shap_values(X_explain.iloc[:200], nsamples=200)


explainer, shap_vals = get_shap_explainer_and_values(
    final_model, best_model_name, X_train, X_test
)

# ── Restrict SHAP output to the same Top 15 features ────────────────────────
top15_feat_names = top15_imp.index.tolist()   # already sorted ascending
top15_indices    = [X_test.columns.tolist().index(f) for f in top15_feat_names]

shap_vals_top15  = shap_vals[:, top15_indices]      # (n_test, 15)
X_test_top15     = X_test[top15_feat_names]

# ── Plot 2: Beeswarm — all test molecules (top 15 features) ─────────────────
fig, ax = plt.subplots(figsize=(10, 7))
shap.summary_plot(shap_vals_top15, X_test_top15,
                  max_display=15, show=False)
ax.set_title(f'SHAP Beeswarm — All Test Molecules (Top 15)\n{best_model_name}',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('shap_beeswarm_top15.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: shap_beeswarm_top15.png")

# ── Plot 3: Bar — mean |SHAP| (top 15 features) ──────────────────────────────
fig, ax = plt.subplots(figsize=(10, 7))
shap.summary_plot(shap_vals_top15, X_test_top15,
                  plot_type='bar', max_display=15, show=False)
ax.set_title(f'SHAP Mean |SHAP| Bar Chart (Top 15)\n{best_model_name}',
             fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('shap_bar_top15.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: shap_bar_top15.png")

# ── Plot 4: Beeswarm — proton sponges only (top 15 features) ─────────────────
if s_mask_test.sum() >= 5:
    shap_sponge = shap_vals_top15[s_mask_test]
    X_test_sponge = X_test_top15.iloc[s_mask_test]
    fig, ax = plt.subplots(figsize=(10, 7))
    shap.summary_plot(shap_sponge, X_test_sponge, max_display=15, show=False)
    ax.set_title(f'SHAP — Proton Sponges Only (Top 15)\n{best_model_name}',
                 fontsize=12, fontweight='bold')
    plt.tight_layout()
    plt.savefig('shap_sponges_top15.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("Saved: shap_sponges_top15.png")

# ── Plot 5: Side-by-side SHAP summary (Beeswarm + Bar) ───────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
fig.suptitle(f'SHAP Feature Importance — {best_model_name} (Test Set, Top 15)',
             fontsize=13, fontweight='bold')

plt.sca(axes[0])
shap.summary_plot(shap_vals_top15, X_test_top15, max_display=15, show=False)
axes[0].set_title('Beeswarm — All Test Molecules', fontweight='bold')

plt.sca(axes[1])
shap.summary_plot(shap_vals_top15, X_test_top15,
                  plot_type='bar', max_display=15, show=False)
axes[1].set_title('Mean |SHAP| — All Test Molecules', fontweight='bold')

plt.tight_layout()
plt.savefig('shap_combined_top15.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: shap_combined_top15.png")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 15 ─ Retrain on ALL data → final production model
# ─────────────────────────────────────────────────────────────────────────────
print(f"\nFitting final production model on ALL {len(df)} molecules...")

X_all_raw = build_feature_matrix(df['mol'], "All data")
X_all_raw = X_all_raw[nonzero_cols]
X_all_raw = X_all_raw.fillna(train_median)
X_all     = X_all_raw.drop(columns=drop_cols)
y_all     = df['PA(kcal/mol)'].values
g_all     = df['is_sponge'].values

prod_model = models[best_model_name]
prod_model.fit(X_all, y_all)

y_pred_all = prod_model.predict(X_all)
print(f"  Production model train R²  : {r2_score(y_all, y_pred_all):.4f}")
print(f"  Production model train MAE : {mean_absolute_error(y_all, y_pred_all):.2f} kcal/mol")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 16 ─ Save all model artifacts
# ─────────────────────────────────────────────────────────────────────────────
joblib.dump(prod_model, 'final_model.pkl', compress=3)
print("Saved: final_model.pkl")

if best_model_name == 'XGBoost':
    prod_model.save_model('final_model.json')
    print("Saved: final_model.json (XGBoost native)")

# Save full predictions on all training data
all_preds_df = pd.DataFrame({
    'SMILES'       : df['can_smi'].values,
    'PA_actual'    : np.round(y_all, 3),
    'PA_predicted' : np.round(y_pred_all, 3),
    'error'        : np.round(y_pred_all - y_all, 3),
    'abs_error'    : np.round(np.abs(y_pred_all - y_all), 3),
    'label'        : ['Proton sponge' if g else 'Normal' for g in g_all],
}).sort_values('abs_error', ascending=False)

all_preds_df.to_csv('all_molecule_predictions.csv', index=False)
print("Saved: all_molecule_predictions.csv")

print(f"\n✓ Training complete. Best model: {best_model_name}")
print("✓ Upload 'final_model.pkl' and 'preprocessing_meta.json' to Colab")
print("  before running PA_PREDICTOR.ipynb")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 17 ─ Download all output files
# ─────────────────────────────────────────────────────────────────────────────
from google.colab import files

critical_files = ['final_model.pkl', 'preprocessing_meta.json']
other_files    = [
    'cv_results.csv',
    'cv_performance.png',
    'test_set_scatter_top3.png',
    'feature_importance_top15.png',
    'feature_importance_top15.csv',
    'shap_beeswarm_top15.png',
    'shap_bar_top15.png',
    'shap_combined_top15.png',
    'shap_sponges_top15.png',
    'all_molecule_predictions.csv',
]
if best_model_name == 'XGBoost':
    other_files.append('final_model.json')

print("★ Downloading critical model artifacts (needed for predictor)...")
for fname in critical_files:
    if os.path.exists(fname):
        files.download(fname)
        print(f"  ✓ {fname}")
    else:
        print(f"  ✗ NOT FOUND: {fname}")

print("\nDownloading analysis outputs...")
for fname in other_files:
    if os.path.exists(fname):
        files.download(fname)
        print(f"  ✓ {fname}")
    else:
        print(f"  – skipped (not generated): {fname}")

print(f"\n✓ Done. Best model: {best_model_name}")
print("  Upload 'final_model.pkl' + 'preprocessing_meta.json' to PA_PREDICTOR notebook.")
```

```python
# =============================================================================
#  PROTON AFFINITY PREDICTOR  —  PA_PREDICTOR.ipynb  (Google Colab)
#
#  REQUIRES (upload to Colab before running):
#    ✓  final_model.pkl          — saved from PA_TRAINER.ipynb
#    ✓  preprocessing_meta.json  — saved from PA_TRAINER.ipynb
#
#  ► No retraining. No dataset needed. Just SMILES → PA prediction.
#  ► To predict new molecules: edit CELL 6 (NEW_SMILES list) and run CELL 6 + 7.
# =============================================================================

# ─────────────────────────────────────────────────────────────────────────────
# CELL 1 ─ Install dependencies  [Run once per session]
# ─────────────────────────────────────────────────────────────────────────────
import subprocess
subprocess.run(["pip", "install", "-q", "rdkit", "xgboost",
                "scikit-learn", "joblib", "pandas", "matplotlib", "openpyxl"],
               check=True)
print("✓ Dependencies installed")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 2 ─ Imports  [Run once per session]
# ─────────────────────────────────────────────────────────────────────────────
import json, warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import joblib

from rdkit import Chem
from rdkit.Chem import Descriptors, MACCSkeys, AllChem, Descriptors3D

warnings.filterwarnings('ignore')
print("✓ Imports ready")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 3 ─ Upload model artifacts  [Run once per session]
# ─────────────────────────────────────────────────────────────────────────────
from google.colab import files

print("Upload 'final_model.pkl' and 'preprocessing_meta.json' ...")
uploaded = files.upload()
print(f"✓ Uploaded: {list(uploaded.keys())}")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 4 ─ Load model and preprocessing metadata  [Run once per session]
# ─────────────────────────────────────────────────────────────────────────────
MODEL_PATH = 'final_model.pkl'
META_PATH  = 'preprocessing_meta.json'

model = joblib.load(MODEL_PATH)
print(f"✓ Model loaded  ({type(model).__name__})")

with open(META_PATH, 'r') as f:
    meta = json.load(f)

NONZERO_COLS  = meta['nonzero_cols']
DROP_COLS     = meta['drop_cols']
TRAIN_MEDIAN  = pd.Series(meta['train_median'])
FINAL_COLUMNS = meta['final_columns']

print(f"✓ Preprocessing metadata loaded: {len(FINAL_COLUMNS)} features")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 5 ─ Descriptor engine & helper functions  [Run once per session]
#   IMPORTANT: This is identical to the trainer — do NOT modify.
# ─────────────────────────────────────────────────────────────────────────────
SMARTS_DEFS = {
    'amidine'     : '[NX3][CX3]=[NX2]',
    'guanidine'   : '[NX3][CX3](=[NX2])[NX3]',
    'peri_diamine': 'c1ccc2c(N)cccc2c1N',
    'bicyclic_N'  : '[N;R2]',
    'phosphazene' : '[NX2]=[PX4]([NX3])([NX3])[NX3]',
    'two_N_close' : '[#7]~[#7]',
    'vinamidine'  : '[NX3]/C=C/C=[NX2]',
}
SMARTS_COMPILED = {k: Chem.MolFromSmarts(v) for k, v in SMARTS_DEFS.items()}


def get_3d_descriptors(mol):
    """3D conformer via ETKDG + MMFF94; falls back to NaN if embedding fails."""
    keys = ['PMI1','PMI2','PMI3','Asphericity','Eccentricity',
            'NPR1','NPR2','RadiusOfGyration','InertialShapeFactor','SpherocityIndex']
    nan_row = {k: np.nan for k in keys}
    try:
        mol_h = Chem.AddHs(mol)
        params = AllChem.ETKDGv3()
        params.randomSeed = 42
        if AllChem.EmbedMolecule(mol_h, params) == -1:
            return nan_row
        res = AllChem.MMFFOptimizeMolecule(mol_h, maxIters=500)
        if res == -1:
            AllChem.UFFOptimizeMolecule(mol_h)
        return {
            'PMI1'               : Descriptors3D.PMI1(mol_h),
            'PMI2'               : Descriptors3D.PMI2(mol_h),
            'PMI3'               : Descriptors3D.PMI3(mol_h),
            'Asphericity'        : Descriptors3D.Asphericity(mol_h),
            'Eccentricity'       : Descriptors3D.Eccentricity(mol_h),
            'NPR1'               : Descriptors3D.NPR1(mol_h),
            'NPR2'               : Descriptors3D.NPR2(mol_h),
            'RadiusOfGyration'   : Descriptors3D.RadiusOfGyration(mol_h),
            'InertialShapeFactor': Descriptors3D.InertialShapeFactor(mol_h),
            'SpherocityIndex'    : Descriptors3D.SpherocityIndex(mol_h),
        }
    except Exception:
        return nan_row


def build_raw_feature_row(mol):
    """Build complete raw feature dict from an RDKit mol object."""
    feat = {}
    for name, fn in Descriptors.descList:
        try:    feat[name] = fn(mol)
        except: feat[name] = np.nan

    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=1024)
    for i in range(1024):
        feat[f'FP_{i}'] = int(fp[i])

    maccs = MACCSkeys.GenMACCSKeys(mol)
    for i in range(167):
        feat[f'MACCS_{i}'] = int(maccs[i])

    for name, pat in SMARTS_COMPILED.items():
        feat[f'has_{name}'] = int(mol.HasSubstructMatch(pat)) if pat else 0

    feat.update(get_3d_descriptors(mol))
    return feat


def smiles_to_features(smiles_str):
    """Full pipeline: SMILES → aligned feature vector ready for model.predict()."""
    smi = str(smiles_str).strip()
    mol = Chem.MolFromSmiles(smi)
    if mol is None:
        return None, None, "INVALID SMILES"

    can_smi = Chem.MolToSmiles(mol)
    mol     = Chem.MolFromSmiles(can_smi)

    raw = build_raw_feature_row(mol)
    row = pd.Series(raw)

    row = row.reindex(NONZERO_COLS)
    row = row.fillna(TRAIN_MEDIAN.reindex(row.index, fill_value=0))
    row = row.drop(index=[c for c in DROP_COLS if c in row.index], errors='ignore')
    row = row.reindex(FINAL_COLUMNS, fill_value=0)

    return row.values.reshape(1, -1), can_smi, "OK"


def classify_pa(pa):
    if pa > 250: return "SUPERBASE (PA > 250)"
    if pa > 240: return "PROTON SPONGE (PA 240–250)"
    if pa > 220: return "STRONG BASE (PA 220–240)"
    return "NORMAL BASE (PA < 220)"


def classify_color(pa):
    if pa > 250: return '#C0392B'
    if pa > 240: return '#E74C3C'
    if pa > 220: return '#E67E22'
    return '#2980B9'

print("✓ Descriptor engine and helpers ready")
print("✓ SESSION SETUP COMPLETE — now edit CELL 6 and run CELL 6 + 7 to predict")

# =============================================================================
# ★★★  PREDICTION CELL  ★★★
# ─────────────────────────────────────────────────────────────────────────────
# CELL 6 ─ Enter SMILES & Predict  [Edit this cell to predict new molecules]
#
#   • Each entry can be:
#       (a) A plain SMILES string:          "CN",
#       (b) A (SMILES, label) tuple:        ("CN", "Methylamine"),
#   • Add or remove rows freely.
#   • Run ONLY this cell (+ CELL 7) for each new prediction batch.
#     Cells 1–5 only need to be run once per session.
# =============================================================================

# ══════════════════════════════════════════════════════════════════════════════
#  ↓↓↓  EDIT YOUR MOLECULES HERE  ↓↓↓
# ══════════════════════════════════════════════════════════════════════════════
NEW_SMILES = [
    # ── Known proton sponge references ──────────────────────────────────────
    (r"CN(C)c1cccc2c(N(C)C)cccc12",              "DMAN"),
    (r"CN(C)c1ccc2c(c1)N(C)C1=CC=CC=C12",        "TMPN"),
    (r"CN(C)c1cccc2cccc(N(C)C)c12",              "DMNA"),

    # ── Ordinary bases (reference) ───────────────────────────────────────────
    ("N",                                         "Ammonia"),
    ("CN",                                        "Methylamine"),
    ("c1ccncc1",                                  "Pyridine"),
    ("N1CCNCC1",                                  "Piperazine"),
    ("NC(=N)N",                                   "Guanidine"),

    # ── Custom / new molecules ───────────────────────────────────────────────
    ("CN(C)c1cccc2c(N(C)C)ccc3ccccc123",          "Mol-1"),
    ("COc1ccc(N(C)C)c2c(N(C)C)ccc(OC)c12",        "Mol-2"),
    ("N(C)C(=N)Nc1cccc2cccc(NC(=N)N(C)C)c12",     "Mol-3"),
    ("N(c1cccc2cccc(N=P(N1CCCC1)(N1CCCC1)N1CCCC1)c12)=P(N1CCCC1)(N1CCCC1)N1CCCC1",
                                                   "Mol-4"),
    (r"CN(C)/C=C(/N=C(\N(C)C)N(C)C)N(C)C",        "Mol-5"),
    (r"CN(C)/C(=C\C(=N)N(C)C)N(C)C",              "Mol-6"),
    (r"CN(C)P(=N/C1=C(\N(C)C)N(C)C1)(N(C)C)N(C)C","Mol-7"),
    ("CN(C)P(=NCC(C)(C)CC(C)(C)C)(N(C)C)N(C)C",   "Mol-8"),
    ("CN(C)P(=NP(=NC(C)(C)C)(N(C)C)N(C)C)(N(C)C)N(C)C",
                                                   "Mol-9"),
    ("CN(C)P(=NP(=NP(=NP(=NC(C)(C)C)(N(C)C)N(C)C)(N(C)C)N(C)C)(N(C)C)N(C)C)(N(C)C)N(C)C",
                                                   "Mol-10"),
    ("CN1c2ccccc2N(C)C1=C",                        "Mol-11"),
    ("Cc1cc(C)cc(C)c1N1CCN(c2c(C)cc(C)cc2C)C1",   "Mol-12"),

    # ── Add your own below ───────────────────────────────────────────────────
    # ("your_SMILES_here", "your_label"),
]
# ══════════════════════════════════════════════════════════════════════════════

# ── Run predictions ───────────────────────────────────────────────────────────
records = []
print("=" * 78)
print(f"  {'#':<4} {'Label':<14} {'Canonical SMILES':<36} {'PA':>8}  Classification")
print("=" * 78)

for i, entry in enumerate(NEW_SMILES, 1):
    if isinstance(entry, tuple):
        smi_raw, label = entry[0], entry[1]
    else:
        smi_raw, label = entry, str(entry)[:14]

    feat_vec, can_smi, status = smiles_to_features(smi_raw)

    if status != "OK":
        print(f"  {i:<4} {label:<14} {'--- ERROR ---':<36} {'---':>8}  {status}")
        records.append({'Label': label, 'Input_SMILES': smi_raw,
                        'Canonical_SMILES': None,
                        'PA_predicted_kcal_mol': None,
                        'Classification': status})
        continue

    pa   = float(model.predict(feat_vec)[0])
    cls  = classify_pa(pa)
    disp = can_smi[:33] + '...' if len(can_smi) > 36 else can_smi
    flag = "  ◄◄◄" if pa > 240 else ""

    print(f"  {i:<4} {label:<14} {disp:<36} {pa:>8.2f}  {cls}{flag}")
    records.append({'Label': label, 'Input_SMILES': smi_raw,
                    'Canonical_SMILES': can_smi,
                    'PA_predicted_kcal_mol': round(pa, 2),
                    'Classification': cls})

print("=" * 78)
print(f"\n  PA reference: Normal < 220 | Strong 220–240 | "
      f"Proton Sponge 240–250 | Superbase > 250  [kcal/mol]")

results_df = pd.DataFrame(records)
results_df.to_csv('new_predictions.csv', index=False)
print(f"\n  ✓ Results saved: new_predictions.csv  ({len(results_df)} molecules)")

# =============================================================================
# CELL 7 ─ Visualise predictions  [Run after CELL 6]
#   Generates three publication-quality plots:
#     Plot A — Horizontal bar chart of predicted PAs
#     Plot B — PA distribution histogram
#     Plot C — Classification pie chart
# =============================================================================

valid = results_df[results_df['PA_predicted_kcal_mol'].notna()].copy()
valid = valid.sort_values('PA_predicted_kcal_mol', ascending=True).reset_index(drop=True)

if len(valid) == 0:
    print("No valid predictions to plot.")
else:
    pa_vals   = valid['PA_predicted_kcal_mol'].values
    labels_ax = valid['Label'].tolist()
    bar_cols  = [classify_color(v) for v in pa_vals]

    # ── Legend patches ────────────────────────────────────────────────────────
    legend_items = [
        mpatches.Patch(color='#2980B9', label='Normal base (< 220 kcal/mol)'),
        mpatches.Patch(color='#E67E22', label='Strong base (220–240 kcal/mol)'),
        mpatches.Patch(color='#E74C3C', label='Proton sponge (240–250 kcal/mol)'),
        mpatches.Patch(color='#C0392B', label='Superbase (> 250 kcal/mol)'),
    ]

    # ════════════════════════════════════════════════════════════════════════
    # PLOT A — Horizontal bar chart (sorted by PA)
    # ════════════════════════════════════════════════════════════════════════
    fig_h = max(5, len(valid) * 0.72)
    fig, ax = plt.subplots(figsize=(13, fig_h))

    bars = ax.barh(labels_ax, pa_vals, color=bar_cols,
                   alpha=0.88, edgecolor='white', linewidth=0.5, height=0.65)

    # Threshold reference lines
    for thresh, col, lbl in [
        (220, '#E67E22', 'Strong base  220'),
        (240, '#E74C3C', 'Proton sponge  240'),
        (250, '#C0392B', 'Superbase  250'),
    ]:
        if pa_vals.min() - 5 < thresh < pa_vals.max() + 10:
            ax.axvline(thresh, color=col, linestyle='--', linewidth=1.5, alpha=0.75)
            ax.text(thresh + 0.3, len(valid) - 0.6, str(thresh),
                    color=col, fontsize=8, va='top', fontweight='bold')

    # Value annotations
    for bar, val in zip(bars, pa_vals):
        ax.text(bar.get_width() + (pa_vals.max() - pa_vals.min()) * 0.01,
                bar.get_y() + bar.get_height() / 2,
                f'{val:.1f}', va='center', fontsize=9, fontweight='bold')

    ax.set_xlabel('Predicted Proton Affinity (kcal/mol)', fontsize=12)
    ax.set_title(f'Predicted Gas-Phase Proton Affinity — New Molecules\n'
                 f'(Model: {type(model).__name__}, sorted by PA)',
                 fontsize=13, fontweight='bold')
    ax.set_xlim(pa_vals.min() - 12, pa_vals.max() + 18)
    ax.legend(handles=legend_items, loc='lower right', fontsize=9, framealpha=0.9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    plt.savefig('predictions_bar_chart.png', dpi=160, bbox_inches='tight')
    plt.show()
    print("Saved: predictions_bar_chart.png")

    # ════════════════════════════════════════════════════════════════════════
    # PLOT B — PA distribution histogram
    # ════════════════════════════════════════════════════════════════════════
    fig, ax = plt.subplots(figsize=(10, 5))

    # Coloured histogram bars per PA range
    bins = np.linspace(pa_vals.min() - 5, pa_vals.max() + 5, 20)
    n_normal  = pa_vals[pa_vals <= 220]
    n_strong  = pa_vals[(pa_vals > 220) & (pa_vals <= 240)]
    n_sponge  = pa_vals[(pa_vals > 240) & (pa_vals <= 250)]
    n_super   = pa_vals[pa_vals > 250]

    for subset, col, lbl in [
        (n_normal, '#2980B9', 'Normal'),
        (n_strong, '#E67E22', 'Strong'),
        (n_sponge, '#E74C3C', 'Proton sponge'),
        (n_super,  '#C0392B', 'Superbase'),
    ]:
        if len(subset) > 0:
            ax.hist(subset, bins=bins, color=col, alpha=0.78, label=lbl,
                    edgecolor='white', linewidth=0.5)

    for thresh, col in [(220, '#E67E22'), (240, '#E74C3C'), (250, '#922B21')]:
        ax.axvline(thresh, color=col, linestyle='--', linewidth=1.5, alpha=0.8)

    ax.set_xlabel('Predicted Proton Affinity (kcal/mol)', fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.set_title('Distribution of Predicted Proton Affinities', fontsize=13, fontweight='bold')
    ax.legend(fontsize=10, framealpha=0.9)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.tight_layout()
    plt.savefig('predictions_histogram.png', dpi=160, bbox_inches='tight')
    plt.show()
    print("Saved: predictions_histogram.png")

    # ════════════════════════════════════════════════════════════════════════
    # PLOT C — Classification pie chart
    # ════════════════════════════════════════════════════════════════════════
    class_counts = valid['Classification'].value_counts()
    color_map = {
        'NORMAL BASE (PA < 220)'      : '#2980B9',
        'STRONG BASE (PA 220–240)'    : '#E67E22',
        'PROTON SPONGE (PA 240–250)'  : '#E74C3C',
        'SUPERBASE (PA > 250)'        : '#C0392B',
    }
    pie_colors = [color_map.get(c, '#7F8C8D') for c in class_counts.index]

    fig, ax = plt.subplots(figsize=(8, 7))
    wedges, texts, autotexts = ax.pie(
        class_counts.values,
        labels=class_counts.index,
        colors=pie_colors,
        autopct='%1.1f%%',
        startangle=140,
        pctdistance=0.82,
        wedgeprops=dict(edgecolor='white', linewidth=1.5),
    )
    for at in autotexts:
        at.set_fontsize(10)
        at.set_fontweight('bold')
    for t in texts:
        t.set_fontsize(9)

    ax.set_title('Classification Distribution of Predicted Molecules',
                 fontsize=13, fontweight='bold', pad=16)

    plt.tight_layout()
    plt.savefig('predictions_pie_chart.png', dpi=160, bbox_inches='tight')
    plt.show()
    print("Saved: predictions_pie_chart.png")

# ─────────────────────────────────────────────────────────────────────────────
# CELL 8 ─ (Optional) Predict from a CSV / Excel file of SMILES
#   File must have a column named 'SMILES' (and optionally 'Label').
# ─────────────────────────────────────────────────────────────────────────────
def predict_from_file(filepath):
    """
    Load a CSV/XLSX with a 'SMILES' column, predict PA for all entries,
    and save results to 'batch_predictions.csv'.

    Usage (uncomment at bottom of cell to run):
        df_out = predict_from_file('my_molecules.csv')
    """
    if filepath.endswith('.csv'):
        df_in = pd.read_csv(filepath)
    elif filepath.endswith(('.xls', '.xlsx')):
        df_in = pd.read_excel(filepath)
    else:
        raise ValueError("File must be .csv or .xlsx")

    assert 'SMILES' in df_in.columns, "File must contain a 'SMILES' column"
    has_label = 'Label' in df_in.columns

    records = []
    n = len(df_in)
    for i, row in df_in.iterrows():
        smi_raw = str(row['SMILES'])
        label   = str(row['Label']) if has_label else f"Mol-{i+1}"
        print(f"  Processing {i+1}/{n}  ({label})...", end='\r')

        feat_vec, can_smi, status = smiles_to_features(smi_raw)
        if status != "OK":
            records.append({'Label': label, 'Input_SMILES': smi_raw,
                            'Canonical_SMILES': None,
                            'PA_predicted_kcal_mol': None,
                            'Classification': status})
            continue

        pa = float(model.predict(feat_vec)[0])
        records.append({'Label': label, 'Input_SMILES': smi_raw,
                        'Canonical_SMILES': can_smi,
                        'PA_predicted_kcal_mol': round(pa, 2),
                        'Classification': classify_pa(pa)})
    print()

    df_out = pd.DataFrame(records)
    df_out.to_csv('batch_predictions.csv', index=False)
    print(f"✓ Done. Saved: batch_predictions.csv ({n} molecules)")
    print("\n  Classification summary:")
    print(df_out['Classification'].value_counts().to_string())
    return df_out


# Uncomment to use:
# df_results = predict_from_file('my_new_library.csv')

# ─────────────────────────────────────────────────────────────────────────────
# CELL 9 ─ Download all result files
# ─────────────────────────────────────────────────────────────────────────────
from google.colab import files

for fname in [
    'new_predictions.csv',
    'predictions_bar_chart.png',
    'predictions_histogram.png',
    'predictions_pie_chart.png',
    'batch_predictions.csv',
]:
    try:
        files.download(fname)
        print(f"✓ Downloaded: {fname}")
    except Exception:
        print(f"  – skipped (not generated): {fname}")
```
